```python
import networkx as nx
import matplotlib.pyplot as plt

graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

# Create a directed graph
G = nx.Graph()

# Add edges to the graph
for i in range(len(graph)):
    for j in graph[i]:
        G.add_edge(i, j)

# Draw the graph
nx.draw(G, with_labels=True, font_weight='bold')

# Show the graph
plt.figure(figsize=(4,4))
plt.show()
```

```python
visited = [False for i in range(len(graph))]
que_ =[]

start_node = 7
que_.append(start_node)

# iteration

while que_:
  a = que_.pop(0)

  if visited[a] == False:

    neis = graph[a]
    visited[a] = True
    print("visit : ", a)

    for n in neis:
      que_.append(n)
```
